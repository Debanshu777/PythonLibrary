### Heart Disease

```python
heart_disease=pd.read_csv("heart_disease.csv")
heart_disease.head(5)
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

### Pyplot vs oo method

```python
heart_disease.head(5)
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head(5)
```

```python
#pyplot
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,6));
```

```python
#oo method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
ax.set(title="CHOL vs AGE")
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
#ax.set_xlim([45,100])
```

```python
## oo method from scretch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter1=plt.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
#customizing the plot
ax.set(title="Cholestrol vs Age",xlabel="Age",ylabel="Cholestrol");

#add legend
ax.legend(*scatter1.legend_elements(),title="Target");
#add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle="--")
```

```python
## oo method from scretch
fig,(ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1,sharex=True)

#plot the data
scatter0=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
scatter1=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
#customizing the plot
ax0.set(title="Cholestrol vs Age",ylabel="Cholestrol");
ax1.set(title="Heart beat vs Age",xlabel="Age",ylabel="Heart beat");

#add legend
ax1.legend(*scatter1.legend_elements(),title="Target");
ax0.legend(*scatter0.legend_elements(),title="Target");
#add horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle="--");
ax1.axhline(over_50['thalach'].mean(),linestyle="--");

#adding figure tilte
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight="bold")
```
